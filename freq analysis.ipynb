```
import matplotlib.pyplot as plt

src = "I HOPE YOU ARE HAVING LOTS OF FUN IN TRYING TO CATCH ME THAT WASNT ME ON THE TV SHOW WHICH BRINGS UP A POINT ABOUT ME I AM NOT AFRAID OF THE GAS CHAMBER BECAUSE IT WILL SEND ME TO PARADICE ALL THE SOONER BECAUSE I NOW HAVE ENOUGH SLAVES TO WORK FOR ME WHERE EVERYONE ELSE HAS NOTHING WHEN THEY REACH PARADICE SO THEY ARE AFRAID OF DEATH I AM NOT AFRAID BECAUSE I KNOW THAT MY NEW LIFE IS LIFE WILL BE AN EASY ONE IN PARADICE"

from collections import defaultdict
freq = defaultdict(int)
for i in "".join(src.split()):
  freq[i] += 1

letters = list(freq.keys())
counts = list(freq.values())
letters, counts = zip(*sorted(freq.items(), key=lambda x: x[1], reverse=True))
plt.figure(figsize=(12, 5))
plt.bar(letters, counts)
plt.xlabel("Letters")
plt.ylabel("Frequency")
plt.title("Letter Frequency Analysis")
plt.show()


```

```
import matplotlib.pyplot as plt
from collections import defaultdict

with open('/content/drive/MyDrive/dracula.txt', encoding='utf-8') as file:
    lines = file.read()

file.close()
src = " ".join(lines)

freq = defaultdict(int)
for i in "".join(src.split()):
  if i.isalnum(): # filter out symbols
    freq[i.upper()] += 1

letters = list(freq.keys())
counts = list(freq.values())
letters, counts = zip(*sorted(freq.items(), key=lambda x: x[1], reverse=True))
plt.figure(figsize=(12, 5))
plt.bar(letters, counts)
plt.xlabel("Letters")
plt.ylabel("Frequency")
plt.title("Letter Frequency Analysis")
plt.show()
```
